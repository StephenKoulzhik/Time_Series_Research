```python
# imports

from utils import *
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
L = 11_000
DELTA = 0.01
N_STEPS = 10
r_0 = 28.0
```

```python
def make_dataset(length, n_steps, delta): 
    data = {}
    # addition of 500 is motivated due to the fact that first 500 observation are dropped as 
    # they are considered to be "calibrating" and have almost nothing to do with the chaotic nature of 
    # Lorenz series
    x, _, _ = Lorentz().generate(0.1, (length + 500 - 1)) #noqa: F405
    x = x[500:]
    data[r_0] = x

    for i in range(1, n_steps):
        r_1 = r_0 - i * delta
        r_2 = r_0 + i * delta
        
        x, _, _ = Lorentz(r=r_1).generate(0.1, (length + 500 - 1)) #noqa: F405
        x = x[500:]
        data[r_1] = x

        x, _, _ = Lorentz(r=r_2).generate(0.1, (length + 500 - 1)) #noqa: F405
        x = x[500:]
        data[r_2] = x 

    return data
```

```python
data = make_dataset(L, N_STEPS, DELTA)
```

```python
data = pd.DataFrame(data)
```
